```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv = pd.read_csv("/content/casasboston.csv")
csv
data = pd.DataFrame(csv)
data
```

```python
df = data[["RM","CRIM","MEDV","TOWN","CHAS"]]

df = data.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS":"RIO CHARLES",
    "RM":"N_HABITACIONES_MEDIO",
    "MEDV":"VALOR_MEDIO",
    "LSTAT":"PCT_CLASE_BAJA"
})
df
```

```python
df.N_HABITACIONES_MEDIO.plot.hist(bins = 7, grid =True, legend = True)
plt.show()
```

```python
df.plot.scatter(x = "VALOR_MEDIO", y = "INDICE_CRIMEN", alpha=0.5)
plt.show()
```

```python
valor_por_ciudad = df.groupby("CIUDAD")["VALOR_MEDIO"].mean()
valor_por_ciudad.head(10).plot.barh()
plt.show()
```

```python
table = pd.crosstab(df["INDICE_CRIMEN"], df["VALOR_MEDIO"])
table
```
